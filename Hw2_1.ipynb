```python
import pandas as pd

data = pd.read_html('https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8', match='Коефіцієнт народжуваності в регіонах України', thousands='.', decimal=',')[0]
```

```python
import numpy as np

data.replace('—', np.NaN, inplace=True)
data
```

```python
data.dtypes
```

```python
data[["2014", "2019"]] = data[["2014", "2019"]].apply(pd.to_numeric)
data.dtypes
```

```python
data.isna().sum()
```

```python
data.drop([27], inplace=True)
data
```

```python
data = data.fillna({"1950":data["1950"].mean(), "1960":data["1960"].mean(), "1970":data["1970"].mean(), "2014":data["2014"].mean(), "2019":data["2019"].mean() })
data
#data.isna().sum()
```

```python
#data[data["2019"] > data["2019"].mean()]["Регіон"]
#data["2019"].mean()
data[data["2019"].gt(data["2019"].mean())]["Регіон"]
```

```python
brate_2019 = data[["Регіон", "2019"]]
brate_2019.plot(kind = 'bar', x="Регіон")
```
