```python
import numpy as np

def get_polynom(coords):
    a = np.ones((len(coords), len(coords)))
    b = np.ones((len(coords),))
    i = 0
    for item in coords:
        for j in range(len(coords), 1, -1):
            a[i, -j] = np.power(item[0], j-1)
        b[i] = item[1]
        i += 1

    return np.linalg.solve(a, b)
```
